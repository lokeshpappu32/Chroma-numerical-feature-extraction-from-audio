```python
import os
import librosa
import numpy as np

# Directory path containing the audio files
directory = 'C:/Users/LOKESH PAPPU/Desktop/NITW/day 11/Mozilla Dataset Adult/Female/'

# Define the number of chroma features
num_chroma_features = 12

# Initialize an empty list to store the extracted chroma features
all_chroma_features = []

# Iterate over all files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.mp3'):
        file_path = os.path.join(directory, filename)
        
        # Load the speech signal
        signal, sr = librosa.load(file_path, sr=None)
        
        # Extract chroma features
        chroma_features = np.mean(librosa.feature.chroma_stft(y=signal, sr=sr, n_chroma=num_chroma_features).T, axis=0)
        
        # Append the chroma features to the list
        all_chroma_features.append(chroma_features)

# Convert the list of chroma features to a numpy array
all_chroma_features = np.array(all_chroma_features)


```
